```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# sample dataframe with multiple coulmns
data = pd.DataFrame({"x":np.random.randn(100),"y":np.random.randn(100),"value":np.random.randn(100)})
cmap="viridis"
alpha=1
plt.figure(figsize=(6,6))
plt.scatter(data["x"],data["y"],c=data["value"],cmap=cmap,alpha=alpha)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scatterplot with colormap")
plt.colorbar(label="value")
plt.show()



```
